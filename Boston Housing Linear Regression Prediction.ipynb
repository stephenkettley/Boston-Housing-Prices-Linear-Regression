```python
#import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#import data

from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston.data.shape
```

```python
boston.feature_names
```

```python
# convert data from nd-array to a dataframe and add feature names to the data

data = pd.DataFrame(boston.data)
data.columns = boston.feature_names

data.head(10)
```

```python
boston.target.shape
```

```python
data['Price'] = boston.target
data.head(10)
```

```python
data.describe()
```

```python
data.info()
```

```python
# getting input and output data and splitting data into training and testing dataset

x = boston.data

y = boston.target

from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2, random_state=0)

print("xtrain shape : ", xtrain.shape)
print("xtest shape  : ", xtest.shape)
print("ytrain shape : ", ytrain.shape)
print("ytest shape  : ", ytest.shape)
```

```python
#apply a linear regression model to the dataset and predicting prices

from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(xtrain,ytrain)

y_pred = regressor.predict(xtest)
```

```python
# plot scatter graph to show the prediction results

plt.scatter(ytest, y_pred, color='g')
plt.xlabel("Actual Price in $1000's")
plt.ylabel('Predicted Prices')
plt.title('True value vs predicted value: Linear Regression')
plt.show()
```

```python
# results of the Linear Regression

from sklearn.metrics import mean_squared_error, mean_absolute_error

mse = mean_squared_error(ytest, y_pred)
mae = mean_absolute_error(ytest, y_pred)

print('Mean squared error is: ', mse)
print('Mean absolute error is: ', mae)
```

The above implies that our model is only 66.55% accurate which means it is not very good for predicting housing prices.
